# 🔍 Automated Data Profiling — Global YouTube Statistics

## Purpose
Before manual cleaning and EDA, this notebook uses `ydata-profiling` to generate a fast, automated statistical profile of the **raw** dataset. The goal is to get an at-a-glance data-quality overview — missing values, data types, distributions, duplicates, and correlations — to confirm and complement the manual findings from the main EDA notebook (`analysis.ipynb`), and to catch anything that manual inspection might have missed.

This is run on the **raw, uncleaned** CSV intentionally — profiling before cleaning gives an honest picture of the data's original quality, rather than a profile of data that's already been fixed.

### Load Raw Data
Loading the original, unprocessed dataset (before any of the cleaning steps applied in the main EDA notebook).

```python
from ydata_profiling import ProfileReport
```

```python
import pandas as pd
```

### Generate Profile Report
`ProfileReport` automatically computes:
- Missing value counts and patterns per column
- Data type detection and warnings (e.g., high cardinality, skewed distributions)
- Correlation matrix across numeric fields
- Duplicate row detection
- Summary statistics for every column

`explorative=True` enables the deeper analysis mode (extra distribution plots and interaction checks) rather than the basic summary.

```python
df = pd.read_csv(r"C:\Users\Lenovo\Krishnaraj singh\Project\Youtube\Data\Global YouTube Statistics.csv",encoding='latin1')
```

### Key Findings from the Profiling Report
- **Missing data:** confirms the same pattern found manually — `Country`, `Abbreviation`, and the demographic columns (Population, Unemployment rate, Latitude/Longitude) are missing together for the same ~9-10% of rows, and `subscribers_for_last_30_days` is missing for ~31% of rows.
- **Duplicates:** `[state whether the report flagged any duplicate rows — likely 0, confirm]`
- **High cardinality columns:** `Youtuber` / `Title` (channel names) are flagged as high-cardinality, as expected — not useful for grouping/aggregation without a Top N filter.
- **Correlations:** `[note the strongest correlation the report shows, e.g., subscribers and video views are strongly positively correlated]`
- **Skew:** `[note which numeric columns are flagged as highly skewed, e.g., subscribers, video views, earnings — expected for this kind of "top creators" dataset]`

The full interactive report is saved as `YoutubeData_profiling_report.html` and can be opened directly in a browser for further drill-down.

```python
profile = ProfileReport(
    df,
    title="Youtube Dataset Profiling Report",
    explorative=True
)
```

```python
profile.to_file("YoutubeData_profiling_report.html")
```

### Conclusion
The automated profile validates the missing-value patterns and skew identified manually in `analysis.ipynb`, giving confidence that the manual cleaning decisions (imputing low-missingness columns, leaving high-missingness/structural gaps as null) were appropriate. This step demonstrates the ability to combine both manual, judgment-driven EDA and automated profiling tools — using each where it's strongest.
